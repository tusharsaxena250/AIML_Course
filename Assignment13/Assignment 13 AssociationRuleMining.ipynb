## Imports

```python
import numpy as np
import pandas as pd
import csv
```

## Working with the csv

```python
words = []
with open("store_data.csv", mode='r') as data:
    raw_data = csv.reader(data, delimiter=',')
    for row in raw_data:
        words.extend(row)
```

```python
words = list(set(words))
print(len(words))
```

```python
words
```

```python
dic = [{i:0 for i in words}]*7501
```

## Dataframe from the csv

```python
df = pd.read_csv("store_data.csv", header=None)
```

## Dataframe of the coded form

```python
dft = pd.DataFrame(dic, columns = words)
dft
```

```python
for i in range(7501):
    row_1 = df.iloc[i]
    for j in row_1:
        dft.iloc[i][j] = 1
```

```python
dft
```

## Stats

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
freq_itemsets = apriori(dft, min_support=0.01, use_colnames=True)
freq_itemsets
```

```python
rules = association_rules(freq_itemsets, metric='confidence', min_threshold=0.5)
rules.head(10)
```

```python
rules[(rules['lift'] > 1.2) & (rules['confidence'] > 0.5)]
```
